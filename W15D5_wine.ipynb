```python
import pandas as pd
import numpy as np
```

```python
file_path = "../Python/beginner datasets/wine.csv"
df = pd.read_csv(file_path)
df
```

```python
# Qual Ã¨ la media di concentrazione alcolica per ogni qualitÃ ? Ci sono differenze? E rispetto alla media totale?
media_totale_perc_alc = df["alcohol"].mean().round(2)
print("\nla media totale della % alcool Ã¨ :", media_totale_perc_alc)
media_qualitÃ _perc_alc = df.groupby(["type", "quality"])["alcohol"].mean().round(2)
print("\nLa concentrazione media di alcool per qualitÃ  Ã¨:", media_qualitÃ _perc_alc)
differenza_rispetto_media_totale = media_qualitÃ _perc_alc - media_totale_perc_alc.round(2)
print("\nDifferenza rispetto alla media totale:", differenza_rispetto_media_totale)
```

```python
# C'Ã¨ differenza nella concentrazione alcolica per vini bianchi e vini rossi?
conc_rosso = df.groupby("type")["alcohol"].mean().round(2)
print("\nLa concentrazione di alcool media per i vini Ã¨:", conc_rosso)
```

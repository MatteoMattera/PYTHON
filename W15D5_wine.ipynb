```python
import pandas as pd
import numpy as np
```

```python
file_path = "../Python/beginner datasets/wine.csv"
df = pd.read_csv(file_path)
df
```

```python
# Qual è la media di concentrazione alcolica per ogni qualità? Ci sono differenze? E rispetto alla media totale?
media_totale_perc_alc = df["alcohol"].mean().round(2)
print("\nla media totale della % alcool è :", media_totale_perc_alc)
media_qualità_perc_alc = df.groupby(["type", "quality"])["alcohol"].mean().round(2)
print("\nLa concentrazione media di alcool per qualità è:", media_qualità_perc_alc)
differenza_rispetto_media_totale = media_qualità_perc_alc - media_totale_perc_alc.round(2)
print("\nDifferenza rispetto alla media totale:", differenza_rispetto_media_totale)
```

```python
# C'è differenza nella concentrazione alcolica per vini bianchi e vini rossi?
conc_rosso = df.groupby("type")["alcohol"].mean().round(2)
print("\nLa concentrazione di alcool media per i vini è:", conc_rosso)
```

```python
# Rifacendo le analisi delle domande precedenti ma per il pH, cambia qualcosa?
media_totale_ph= df["pH"].mean().round(2)
print("\nla media totale del ph :", media_totale_ph)
media_qualità_ph = df.groupby(["type", "quality"])["pH"].mean().round(2)
print("\nil ph medio per qualità è:", media_qualità_ph)
#differenza_rispetto_media_totale_ph = media_qualità_ph - media_totale_ph.round(2)
#print("\nDifferenza rispetto alla media totale:", differenza_rispetto_media_totale_ph)
```

```python
# E per i solfati?
media_totale_solfati= df["sulphates"].mean().round(2)
print("\nla media totale del ph :", media_totale_solfati)
media_qualità_solfati = df.groupby(["type","quality"])["sulphates"].mean().round(2)
print("\nil ph medio per qualità è:", media_qualità_solfati)
differenza_rispetto_media_totale_ph = media_qualità_solfati - media_totale_ph.round(2)
print("\nDifferenza rispetto alla media totale:", differenza_rispetto_media_totale_ph)
```

```python
# Quali sono le medie di charges rispetto a region? Ci sono differenze significative?
```

```python
medie_per_regione = df.groupby("region")["charges"].mean()
print(medie_per_regione)
```

```python
# E rispetto a smoker? 
medie_per_smoker = df.groupby("smoker")["charges"].mean()
print(medie_per_smoker)
```

```python
#E per sex?
medie_per_sex = df.groupby("sex")["charges"].mean()
print(medie_per_sex)
```

```python
#Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?
df.groupby('bmi').describe().transpose() # il transpose ribalta le colonne sulle righe


```

```python
quartili_bmi = pd.qcut(df["bmi"], q=4)

dati = df.groupby(quartili_bmi)["charges"].agg(["min", "mean", "max"]).transpose()
print(dati)
```
